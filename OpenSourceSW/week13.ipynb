```python
import csv
import matplotlib.pyplot as plt
f=open('seoul.csv')
data=csv.reader(f)
next(data)
high=[]
low=[]
for row in data:
    # 최고 기온
    if row[-1] != '': # 데이터 누락 확인
        if row[0].split('-')[1]=='11' and row[0].split('-')[2]=='17':
            high.append(float(row[-1]))
    # 최저 기온
    if row[-2] != '':
        if row[0].split('-')[1]=='11' and row[0].split('-')[2]=='17':
            low.append(float(row[-2]))

# 한글 쓸때
plt.rc('font', family='NanumGothic')
plt.title('11월 17일의 최고기온과 최저기온')
plt.plot(high, color='r',label='최고기온')
plt.plot(low, color='b', label='최저기온')
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
f=open('seoul.csv')
data=csv.reader(f)
next(data)
jan=[]
aug=[]
for row in data:
    if row[-1] != '':
        if row[0].split('-')[1]=='01':
            jan.append(float(row[-1]))
        elif row[0].split('-')[1]=='08':
            aug.append(float(row[-1]))
plt.rc('font', family='NanumGothic')
plt.title('서울의 1월과 8월 최고 기온')
plt.hist(jan, bins=100, color='b', label='1월')
plt.hist(aug, bins=100, color='r', label='8월')
plt.legend()
plt.show()
```
