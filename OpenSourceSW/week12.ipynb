```python
import csv
# 읽기 모드, 인코딩은 유니코드
f=open('seoul.csv','r',encoding='cp949')
# 구분자는 컴마
data=csv.reader(f,delimiter=',')
# 헤더는 필요없는 정보이므로 next
header=next(data)
#print(header)

# 일교차가 가장 큰 날, 최저 기온이 가장 낮은 날, 최고 기온이 가장 높은 날
tems, days=[0]*3, ['']*3
for row in data:
    if len(row) != 5:
        continue
    low, high=float(row[3]),float(row[4])
    differ=high-low
    if differ>tems[0]:
        tems[0]=differ
        days[0]=row[0]
    if low<tems[1]:
        tems[1]=low
        days[1]=row[0]
    if high>tems[2]:
        tems[2]=high
        days[2]=row[0]
# 출력
print("일교차가 가장 큰 날")
print(days[0])
print("최저기온이  가장 낮은 날")
print(days[1])
print("최고기온이 가장 높은 날")
print(days[2])
```

```python
import csv
from collections import defaultdict
f=open('subwayfee.csv','r')
data=csv.reader(f,delimiter=',')
header=next(data)

# 각 노선별로 지하철역과 승객 수. 노선이 key값
routes=defaultdict(list)
# 튜플(이용객,역)형태로 리스트에 저장
for row in data:
    # 1,123,456의 컴마를 제거하여 1123456을 int로 캐스팅
    passenger=int(row[4].replace(',',''))+int(row[5].replace(',',''))
    routes[row[1]].append((passenger,row[3]))
# 내림차순 정렬
for r in routes.keys():
    print("\n======노선:{0:s}======".format(r))
    routes[r].sort(reverse=True)
    print("이용객이 가장 많은 역 순위")
    for i in range(3):
        print("{0:d}. {1:s}({2:d})".format(i+1,routes[r][i][1], routes[r][i][0]))
    print("이용객이 가장 적은 역 순위")
    for i in range(3):
        print("{0:d}. {1:s}({2:d})"
              .format(i+1,routes[r][len(routes[r])-i-1][1], routes[r][len(routes[r])-i-1][0]))
```
