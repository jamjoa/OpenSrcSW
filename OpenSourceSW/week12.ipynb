```python
import csv
# 읽기 모드, 인코딩은 유니코드
f=open('seoul.csv','r',encoding='cp949')
# 구분자는 컴마
data=csv.reader(f,delimiter=',')
# 헤더는 필요없는 정보이므로 next
header=next(data)
#print(header)

# 일교차가 가장 큰 날, 최저 기온이 가장 낮은 날, 최고 기온이 가장 높은 날
tems, days=[0]*3, ['']*3
for row in data:
    if len(row) != 5:
        continue
    low, high=float(row[3]),float(row[4])
    differ=high-low
    if differ>tems[0]:
        tems[0]=differ
        days[0]=row[0]
    if low<tems[1]:
        tems[1]=low
        days[1]=row[0]
    if high>tems[2]:
        tems[2]=high
        days[2]=row[0]
# 출력
print("일교차가 가장 큰 날")
print(days[0])
print("최저기온이  가장 낮은 날")
print(days[1])
print("최고기온이 가장 높은 날")
print(days[2])
```

```python
import csv
f=open('subway.csv','r')
data=csv.reader(f,delimiter=',')
header=next(data)
# print(header)
stations=[]
for row in data:
    passenger=int(row[4].replace(',',''))+int(row[5].replace(',',''))
    stations.append((passenger,row[3]))
stations.sort(reverse=True)
print("이용객이 가장 많은 역 순위")
for i in range(3):
    print("{0:d}. {1:s}({2:d})".format(i+1,stations[i][1], stations[i][0]))
print("이용객이 가장 적은 역 순위")
for i in range(3):
    print("{0:d}. {1:s}({2:d})"
          .format(i+1,stations[len(stations)-i-1][1], stations[len(stations)-i-1][0]))
```
