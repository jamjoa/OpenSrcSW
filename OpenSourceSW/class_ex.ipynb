```python
# 실습 8 사칙 연산 클래스
class Calc:    
    def print_result(self,a,b,oper,result):
        print("{0:d} {1:s} {2:d} = {3:d} 입니다."
             .format(a,oper,b,result))
        
    def sum(self,a,b):
        result=a+b
        oper='+'
        self.print_result(a,b,oper,result)
    
    def sub(self,a,b):
        result=a-b
        oper='-'
        self.print_result(a,b,oper,result)
        
    def multi(self,a,b):
        result=a*b
        oper='*'
        self.print_result(a,b,oper,result)
        
    def divi(self,a,b):
        result=int(a/b)
        oper='/'
        self.print_result(a,b,oper,result)
    
if __name__ == "__main__":
    a=int(input())
    b=int(input())
    test=Calc()
    test.sum(a,b)
    test.sub(a,b)
    test.multi(a,b)
    test.divi(a,b)
```
