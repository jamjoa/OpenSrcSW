```python
# 최대값 구하기
input_string=input("Numbers?")
input_list=input_string.split()
input_list=[int(i) for i in input_list]
print("Max =",max(input_list))
```

```python
# 암호화/복호화
_input=input("문자열을 입력하세요 :")
_input.strip()
input_list=_input.split()
print("입력받은 문자열 :",' '.join(input_list))
input_list.reverse()
encrypt=[]
for s in input_list:
    r=""
    for c in s:
        code=ord(c)
        r+=chr(code+1)
    encrypt.append(r)
print("암호화 결과 :",' '.join(encrypt))
decrypt=[]
for s in encrypt:
    r=""
    for c in s:
        code=ord(c)
        r+=chr(code-1)
    decrypt.append(r)
decrypt.reverse()
print("복호화 결과: ",' '.join(decrypt))
```

```python
# 전화번호부 실습
telbook={}
print("10명의 이름과 전화번호를 입력하세요")
for i in range(3):
    s=input().split()
    telbook[s[0]]=s[1]
while True:
    m=input("1) 이름으로 검색 2) 번호로 검색 3) 종료 =>")
    if m=='1':
        print("현재 이름 목록 :",'/'.join(telbook.keys()))
        s=input("이름 :")
        print("전화번호 :", telbook[s])
    elif m=='2':
        print("현재 번호 목록 :",'/'.join(telbook.values()))
        s=input("전화번호 :")
        ans=""
        for k,v in telbook.items():
            if v==s :
                ans=k
                break
        print("이름 :", ans)
    elif m=='3':
        break
    else:
        print("오류! 다시 입력해주세요")
```
