```python
# 네이버 검색 API예제는 블로그를 비롯 전문자료까지 호출방법이 동일하므로 blog검색만 대표로 예제를 올렸습니다.
# 네이버 검색 Open API 예제 - 블로그 검색
import os
import sys
import requests
import urllib.parse
client_id = "CSawgvaLFzZ8Ss7Zmh38"
client_secret = "s3UwATTbdx"
keyword=input("검색어를 입력하세요: ")
url = "https://openapi.naver.com/v1/search/movie?query=" + keyword # json 결과
result = requests.get(urllib.parse.urlparse(url).geturl(),
                     headers={"X-Naver-Client-Id":client_id,
                            "X-Naver-Client-Secret":client_secret})

if not result.ok:
    print("Error Code:" , result.status_code)
else:
    # print(result.text)
    json_obj=result.json()
    items=json_obj.get('items')
    count=0
    for i in items:
        print("=================={0:d}번째 영화==================".format(count+1))
        for key in i.keys():
            if key in ["title","subtitle","director","actor","userRating"]:
                if i[key] != '':
                    print("{0:s}:\t{1:s}".format(key,i[key]))
        print()
        count+=1
        
```
