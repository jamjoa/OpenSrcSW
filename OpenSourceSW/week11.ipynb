```python
class Stack:
    s=[]
    def pop(self):
        return self.s.pop()
    def push(self,num):
        self.s.append(num)
    def size(self):
        return len(self.s)
    def top(self):
        return self.s[self.size()-1]
    def empty(self):
        return len(self.s)==0

def in_to_pre(inorder):
    # 중위 표기식에서 후위 표기식으로 변환하여 return
    postorder=[]
    stack=Stack()
    opers=['+','-','*','/']
    for a in inorder:
        if a in opers:
            # 스택이 비어있을 경우
            if stack.empty():
                stack.push(a)
            # a의 우선순위 더 큼
            elif a in '*/':
                if stack.top() in '+-':
                    stack.push(a)
            # 그 외
            else:
                while not stack.empty():
                    postorder.append(stack.pop())
                stack.push(a)
        else:
            postorder.append(a)
    while not stack.empty():
        postorder.append(stack.pop())
    return postorder

def calc_postorder(postorder):
    ans=[]
    for i in postorder:
        if i not in '+-*/':
            ans.append(i)
        else:
            b=int(ans.pop())
            a=int(ans.pop())
            if i=='+':
                ans.append(a+b)
            elif i=='-':
                ans.append(a-b)
            elif i=='*':
                ans.append(a*b)
            elif i=='/':
                ans.append(int(a/b))
    return ans[0]
            
if __name__=="__main__":
    user_input=input("계산식 입력: ")
    inorder=user_input.split()
    postorder=in_to_pre(inorder)
    ans=calc_postorder(postorder)
    print("계산식 출력: {0:s}={1:d}".format(''.join(inputs),ans))
    
```
