```python
class Point:
    num1,num2=0,0
    def __init__(self,n1, n2):
        self.num1=n1
        self.num2=n2
    def __add__(self, point2):
        return Point(self.num1+point2.num1, self.num2+point2.num2)
    def show(self):
        print("({0:d}, {1:d})".format(self.num1,self.num2))
        
if __name__=="__main__":
    a=Point(1,2)
    a.show()
    b=Point(3,4)
    b.show()
    a=a+b
    a.show()
    
```
